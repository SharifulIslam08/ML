# We will be using apriori algorithm from package ----mlxtend---


```python
#Apriori is array based algorithm
#It works on BFS
#Candidate generation is very slow
#As BFS it takes a lot of memory
```

```python
# Importing all necessary packages for apriori algorithm
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
data=pd.read_csv('Online Retail.csv',encoding= 'unicode_escape')
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
#As Description has many null values we will drop all those values
```

```python
data.shape
```

```python
data1=data.dropna()
```

```python
data1.InvoiceDate=data1.InvoiceDate.astype('str')
```

```python
# We will be dividing date into month,day and year
```

```python
data['Month']=0
```

```python
data.head(10)
```

```python
data1['year'] = pd.DatetimeIndex(data1['InvoiceDate']).year
```

```python
data1['Month'] = pd.DatetimeIndex(data1['InvoiceDate']).month
```

```python
data1['Day'] = pd.DatetimeIndex(data1['InvoiceDate']).day
```

```python
data1.head(10)
```

```python
data1=data1.drop('InvoiceDate',axis=1)
```

```python
data1.head(5)
```

```python
#Creating new column for Total amount spend by one customer
data1['Amount_spent']=data1['Quantity']*data1['UnitPrice']
```

```python
data1.head()
```

```python
data1['Amount_spent'].describe()
```

```python
#Here we can see we have some Amount spent values in negative.We need to remove all those
```

```python
#deleting negative values
data1.Amount_spent.astype('float')
indexNames = data1[ data1['Amount_spent'] <0.0 ].index
 
# Delete these row indexes from dataFrame
data1.drop(indexNames , inplace=True)
```

```python
data1['Amount_spent'].describe()
```

```python
data1['Quantity']=data1['Quantity'].astype('int')
```

```python
#Making invoiceNo as index and all columns describe one item in the dataset
#Filled all the empty space with 0
basket=(data1[data1['Country']=='United Kingdom'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket.info()
```

```python
basket=basket.astype('int64')
```

```python
basket.info()
```

```python
basket=basket.astype('bool')
```

```python
basket
```

```python
#!pip install --no-binary :all: mlxtend
```

```python

frequent_itemsets=apriori(basket,min_support=0.07,use_colnames='True')
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.7)
rules.head()
```

```python
#As we can see for minimum_support=0.07 no items are matching,an empty list is generated
#So we will try by changing the min_support value to 0.03
```

```python

frequent_itemsets=apriori(basket,min_support=0.03,use_colnames='True')
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.3)
rules.head()
```

```python
#So we will try by changing the min_support value to 0.02
```

```python
#The confidence metric is commonly used to measure how much more often the antecedent and consequent of a rule A->C occur together than we would expect if they were statistically independent. 
#If A and C are independent, the Lift score will be exactly 1.

frequent_itemsets=apriori(basket,use_colnames='True',min_support=0.02)
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.02)
rules.head()
```

```python
frequent_itemsets=apriori(basket,use_colnames='True',min_support=0.01)
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.4)
rules.head()
```
